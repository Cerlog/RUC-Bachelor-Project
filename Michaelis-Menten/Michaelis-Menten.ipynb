```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update({'text.usetex': True}) # import to make plots nicer
%config InlineBackend.figure_format = 'svg' # import to make plots nicer

# Set global font size for title, x-label, and y-label
plt.rcParams['axes.titlesize'] = 20
plt.rcParams['axes.labelsize'] = 16

# Set global font size for x and y tick labels
plt.rcParams['xtick.labelsize'] = 16
plt.rcParams['ytick.labelsize'] = 16

# Set global font size for the legend
plt.rcParams['legend.fontsize'] = 16

plt.rcParams['figure.dpi'] = 300
# Set global figure size (width, height) in inches
plt.rcParams['figure.figsize'] = (12, 8)
```

```python
def michaelis_menten(S, V_max, K_m):
    return (V_max * S) / (K_m + S)

# Parameters
V_max = 10  # Maximum reaction velocity
K_m = 5     # Michaelis constant
substrate_concentrations = np.linspace(0, 70, 1000)  # Substrate concentration range

# Calculate initial velocities for the substrate concentrations
initial_velocities = michaelis_menten(substrate_concentrations, V_max, K_m)




# Plot V_max
plt.axhline(V_max, linestyle='--', color='red', label=f'V_max = {V_max}')
plt.text(60/2, V_max + 0.5, r'$V_{max} = k2[E]_{total}$', color='red', fontsize=18)

# Plot 1/2 V_max
half_v_max = V_max / 2
plt.plot([0, K_m], [half_v_max, half_v_max], linestyle='--', color='green', label=f'1/2 V_max = {half_v_max}')
plt.text(1 + 5, half_v_max, r'$\frac{1}{2}V_{max}$', color='green', fontsize=18)

# Plot vertical line at K_m, but only below the curve
plt.plot([K_m, K_m], [0, half_v_max], linestyle='--', color='green')

# Plot the Michaelis-Menten curve
plt.plot(substrate_concentrations, initial_velocities, label='Michaelis-Menten curve')
# Customize plot
plt.xlabel('Substrate concentration, [S] (mM)')
plt.ylabel(r'Initial velocity, $V_0 (\mu M/min)$')
plt.title('Effect of substrate concentration on the initial velocity of an enyzme-catalyzed reaction')
plt.xlim(0.0,60)
plt.ylim(0.0,12)
#plt.legend(fontsize=16)
plt.savefig('Michaelis-Menten.png', dpi=300)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams.update(mpl.rcParamsDefault)
plt.rcParams.update({'text.usetex': True})
%config InlineBackend.figure_format = 'svg'
from matplotlib.lines import Line2D

# Set global font size for title, x-label, and y-label
plt.rcParams['axes.titlesize'] = 20

plt.rcParams['axes.labelsize'] = 16


# Set global font size for x and y tick labels
plt.rcParams['xtick.labelsize'] = 16
plt.rcParams['ytick.labelsize'] = 16

# Set global font size for the legend
plt.rcParams['legend.fontsize'] = 16
plt.rcParams['figure.figsize'] =  (8, 6)
```

```python
def CAMP(t, x, k1, k3, k2, r1, r2, Km1, Km2, Dt, W0, PDEPt):
    cAMP, ACpt = x
    dACPt = r1*cAMP*((W0 - ACpt)/(Km1 + cAMP + (W0 - ACpt))) - r2*Dt*ACpt/(Km2 + ACpt + Dt)
    dcAMP = k1*ACpt - (k2*PDEPt + k3)*cAMP
    return np.array([dACPt, dcAMP])

W0 = 10
W1 = 6.54
k = 0.4
k1, k3, k2 = 4.32/k, 0.58/k, 3.73/k
r1, r2, r3, r4 = 0.98/k, 4.48/k, 0.1/k, 0.14/k
Km1, Km2, Km3, Km4 = 2.48, 3.34, 9.24, 7.24
Dt, Et =  0.58, 1.6
```

```python
def plot_nullclines(k1, k3, k2, r1, r2, Km1, Km2, Dt, W0, PDEPt):
    cAMP_vals = np.linspace(-30, 30, 1000)
    dACpt_vals = np.linspace(-30, 30, 1000)
    cAMP_grid, dACpt_grid = np.meshgrid(cAMP_vals, dACpt_vals)

    t = 0
    x = np.array([cAMP_grid, dACpt_grid])
    nullclines = CAMP(t, x, k1, k3, k2, r1, r2, Km1, Km2, Dt, W0, PDEPt)
    plt.figure(figsize=(8, 6))
    plt.scatter(0, 0, color='black', marker='o', label='Fixed point (0,0,0)', zorder=3)
    plt.contour(cAMP_grid, dACpt_grid, nullclines[0], levels=[0], colors='#1f77b4', linestyles='solid')
    plt.grid(linestyle=':', color='lightgray')
    plt.contour(cAMP_grid, dACpt_grid, nullclines[1], levels=[0], colors='green', linestyles='solid')
    plt.xlabel('cAMP')
    plt.ylabel('ACpt')
        # Create custom legend handles and labels
    custom_lines = [Line2D([0], [0], color='#1f77b4', lw=2),
                    Line2D([0], [0], color='green', lw=2),
                    Line2D([0], [0], color='black', marker='o', linestyle='None')]
    #plt.legend(custom_lines, ['cAMP', 'ACp', 'Fixed point (0.70,0.69,1.00)'])
    plt.legend(custom_lines, ['cAMP', 'ACpt', 'Fixed point (0,0,0)'], loc='upper left')
    plt.title(f'ACpt and cAMP nullclines with fixed PDEp at {PDEPt}')
    plt.ylim(-3, 5)
    plt.xlim(-5,5)
    plt.savefig(f'Nullclines_for_ACpt_and_cAMP_with_PDEp_fixed_at_{PDEPt}_big.png', dpi=300)
    plt.show()
```

```python
plot_nullclines(k1, k3, k2, r1, r2, Km1, Km2, Dt, W0, 0)
```

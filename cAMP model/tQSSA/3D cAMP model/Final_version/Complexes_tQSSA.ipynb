```python
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams.update({'text.usetex': True})
%config InlineBackend.figure_format = 'svg'


# Set global font size for title, x-label, and y-label
plt.rcParams['axes.titlesize'] = 20
plt.rcParams['axes.labelsize'] = 16


# Set global font size for x and y tick labels
plt.rcParams['xtick.labelsize'] = 16
plt.rcParams['ytick.labelsize'] = 16

# Set global font size for the legend
plt.rcParams['legend.fontsize'] = 16
plt.rcParams['figure.figsize'] = (8, 6)
```

```python
k = 0.4 #time scaling
W0 = 10 ; W1 = 6.54
k1, k3, k2 = 4.32/k, 0.58/k, 3.73/k
r1, r2, r3, r4 = 0.98/k, 4.48/k, 0.1/k, 0.14/k
Km1, Km2, Km3, Km4 = 2.48, 3.34, 9.24, 7.24
Dt, Et =  0.58, 1.6
```

```python
cAMP_t1 = np.linspace(0, 100, 100)
AC_t1 = np.linspace(0, 100, 100)
cAMP_2 = np.linspace(0, 100, 100)
PDE_t1 = np.linspace(0, 100, 100)

ACcAMP = cAMP_t1 * AC_t1 / Km1 + cAMP_t1 + AC_t1
PDEcAMP = cAMP_2 * PDE_t1 / Km2 + cAMP_2 + PDE_t1

plt.figure(figsize=(8, 6))
plt.plot(cAMP_t1, ACcAMP, label = r'$[ACcAMP]$', linestyle = '--', linewidth = 2.5, markersize=3)
plt.plot(cAMP_2, PDEcAMP, label = r'$[PDE cAMP]$', linestyle = '--', linewidth = 2.5, markersize=3)
plt.xlabel(r'$cAMP_{t1}$, $AC_{t1}$, $cAMP_{t2}$ and $PDE_{t1}$')
plt.title(r'Plot of $[ACcAMP]$ and $[PDE cAMP]$')
plt.ylabel('$[ACcAMP]$ and  $[PDE cAMP]$')
plt.legend()
plt.savefig('ACcAMP and PDEcAMP.png', dpi=300)
plt.show()
```

```python
ACpt = np.linspace(0, 100, 100)
PDEpt = np.linspace(0, 100, 100)
ACpD = Dt * ACpt / Km2 + ACpt + Dt
PDEpE = Et * PDEpt / Km4 + PDEpt + Et

plt.figure(figsize=(8, 6))
plt.plot(ACpt, ACpD, label=r'$[AC_pD]$', linestyle='--', linewidth=2.5, markersize=3, color='indigo')
plt.plot(PDEpt, PDEpE, label=r'$[PDE_pE]$', linestyle='--', linewidth=2.5, markersize=3, color='lightcoral')
plt.title(r'Plot of $[AC_pD]$ and $[PDE_pE]$')
plt.ylabel('$[AC_pD]$ and  $[PDE_pE]$')
plt.xlabel(r'$AC_{pt}$ and $PDE_{pt}$')
plt.legend()
plt.savefig('ACpD and PDEpD.png', dpi=300)
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np

mpl.rcParams.update(mpl.rcParamsDefault)
plt.rcParams.update({'text.usetex': True})
%config InlineBackend.figure_format = 'svg'
from scipy.integrate import solve_ivp

plt.rcParams['axes.titlesize'] = 20
plt.rcParams['axes.labelsize'] = 16

plt.rcParams['xtick.labelsize'] = 16
plt.rcParams['ytick.labelsize'] = 16

plt.rcParams['legend.fontsize'] = 16
plt.rcParams['figure.figsize'] = (8, 6)
```

```python
def cAMP_PDEp_ACp(t, x):
    cAMP, PDEpt, ACpt = x

    dcAMP = k1*ACpt - (k2*PDEpt + k3)*cAMP
    dACPt = r1*cAMP*((W0 - ACpt)/(Km1 + cAMP + (W0 - ACpt))) - r2*Dt*ACpt/(Km2 + ACpt + Dt)
    dPDEpt = r3*cAMP*((W1 - PDEpt)/(Km3 + cAMP + (W1 - PDEpt))) - r4*Et*(PDEpt)/(Km4 + PDEpt + Et)
    return np.array([dcAMP, dPDEpt, dACPt])
W0 = 5
k = 0.4 #time scaling
W1 = 6.54
k1, k3, k2 = 4.32/k, 0.58/k, 3.73/k
r1, r2, r3, r4 = 0.98/k, 4.48/k, 0.1/k, 0.14/k
Km1, Km2, Km3, Km4 = 2.48, 3.34, 9.24, 7.24
Dt, Et =  0.58, 1.6

x0_values = [[0.33, 0.33, 0.33], [0.5, 0.5, 0.5], [1.0, 1.0, 1.0]]

a, b = 0, 1000
t = np.linspace(a, b, 5000)
angle = 90
angle2 = 150

fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')
plt.subplots_adjust(top=0.85)

for x0 in x0_values:
    sol = solve_ivp(cAMP_PDEp_ACp, [a, b], x0, t_eval=t)
    ax.plot(sol.y[0], sol.y[1], sol.y[2], label=rf"$cAMP_0={x0[0]:.2f}$, $PDEp_0={x0[1]:.2f}$, $ACp_0={x0[2]:.2f}$")

ax.set_title('Limit cycle of the cAMP-PDEp-ACp oscillations with tQSSA', y=1)
ax.set_xlabel('cAMP', fontsize=16, labelpad=5)
ax.set_ylabel('PDEp', fontsize=16, labelpad=5)
ax.set_zlabel('ACp', fontsize=16, labelpad=-1)
ax.view_init(15,angle)
ax.set_box_aspect([1, 1, 1])
ax.grid(True, linestyle=':', color='lightgray')

ax.legend(fontsize=16, bbox_to_anchor=(0.5, 0.9))

plt.savefig(f'3DcAMP_PDEp_ACp_limit_cycle_sQSSA_{W0}_{angle2}.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
# if getting error, please mute the three lines below
mpl.rcParams.update(mpl.rcParamsDefault)
plt.rcParams.update({'text.usetex': True}) # import to make plots nicer
%config InlineBackend.figure_format = 'svg' # import to make plots nicer
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt




# Set global font size for title, x-label, and y-label
plt.rcParams['axes.titlesize'] = 20
plt.rcParams['axes.labelsize'] = 16


# Set global font size for x and y tick labels
plt.rcParams['xtick.labelsize'] = 16
plt.rcParams['ytick.labelsize'] = 16

# Set global font size for the legend
plt.rcParams['legend.fontsize'] = 16
plt.rcParams['figure.figsize'] = (8, 6)
```

```python
def cAMP_PDEp_ACp(t, x):
    cAMP = x[0]
    PDEp = x[1]
    ACp = x[2]

    dPDEp  = r3 * cAMP * ((PDEt - PDEp) / Km3) - r4 * Et * PDEp / (Km4 + PDEp)
    dACp = r1 * cAMP * ((ACt-ACp) / Km1) - r2 * Dt * ACp / (Km2 + ACp)
    dcAMP = k1 * ACp - (k3 + k2 * PDEp) * cAMP
    return [dcAMP, dPDEp, dACp]

k = 0.9
k1, k3, k2 = 4.32/k, 0.58/k, 3.73/k
r1, r2, r3, r4 = 0.98/k, 4.48/k, 0.1/k, 0.14/k
Km1, Km2, Km3, Km4 = 2.48, 3.34, 9.24, 7.24
Dt, Et =  2.12, 1.6
ACt = 5.390539053905391
#ACt = 4.390539053905391
#ACt = 6.390539053905391
PDEt = 4.64

x0_values = [[0.33, 0.33, 0.33], [0.5, 0.5, 0.5], [1.0, 1.0, 1.0]]

a, b = 0, 1000
t = np.linspace(a, b, 5000)
angle = 90
angle2 = 150

fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')
plt.subplots_adjust(top=0.85)

for x0 in x0_values:
    sol = solve_ivp(cAMP_PDEp_ACp, [a, b], x0, t_eval=t)
    ax.plot(sol.y[0], sol.y[1], sol.y[2], label=rf"$cAMP_0={x0[0]:.2f}$, $PDEp_0={x0[1]:.2f}$, $ACp_0={x0[2]:.2f}$")

ax.set_title('Limit cycle of the cAMP-PDEp-ACp oscillations with sQSSA', y=1)
ax.set_xlabel('cAMP', fontsize=16, labelpad=5)
ax.set_ylabel('PDEp', fontsize=16, labelpad=5)
ax.set_zlabel('ACp', fontsize=16, labelpad=-1)
ax.view_init(15,90)
ax.set_box_aspect([1, 1, 1])
ax.grid(True, linestyle=':', color='lightgray')

ax.legend(fontsize=16, bbox_to_anchor=(0.5, 0.9))

plt.savefig(f'3DcAMP_PDEp_ACp_limit_cycle_sQSSA_{ACt}_{angle2}.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
mpl.rcParams.update(mpl.rcParamsDefault)
plt.rcParams.update({'text.usetex': True}) # import to make plots nicer
%config InlineBackend.figure_format = 'svg' # import to make plots nicer

# Set global font size for title, x-label, and y-label
plt.rcParams['axes.titlesize'] = 20
plt.rcParams['axes.labelsize'] = 16

# Set global font size for x and y tick labels
plt.rcParams['xtick.labelsize'] = 16
plt.rcParams['ytick.labelsize'] = 16

# Set global font size for the legend
plt.rcParams['legend.fontsize'] = 16
plt.rcParams['figure.figsize'] = (8, 6)
```

```python
k = 0.9
k1, k3, k2 = 4.32/k, 0.58/k, 3.73/k
r1, r2, r3, r4 = 0.98/k, 4.48/k, 0.1/k, 0.14/k
Km1, Km2, Km3, Km4 = 2.48, 3.34, 9.24, 7.24
Dt, Et =  2.12, 1.6
ACt = 6.68
PDEt = 4.64

PDE_p = np.linspace(0, 100, 100)
AC_p = np.linspace(0, 100, 100)

PDE_pE = Et * PDE_p / (Km4 + PDE_p)
AC_pD = Dt * AC_p / (Km2 + AC_p)

# Plot the concentrations of the complexes
plt.figure(figsize=(8,6))
plt.plot(PDE_p, PDE_pE, label=r'$[PDE_pE]$', zorder=2, linestyle='--', color='crimson', linewidth=2.5, markersize=3)
plt.plot(AC_p, AC_pD, label=r'$[AC_pD]$', zorder=1, linestyle='-.', color='dodgerblue', linewidth=2.5, markersize=3)
plt.legend()
plt.xlabel(r'$[PDE_p]$ and $[AC_p]$')
plt.ylabel('$[PDE_pE]$ and $[AC_pD]$')
plt.title(r'Plot of $[PDE_pE]$ and $[AC_pD]$')
plt.grid(True, linestyle=':', color='lightgray')
plt.savefig('PDE_pE_AC_pD_complex.png', dpi=300)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
k = 0.9
k1, k3, k2 = 4.32/k, 0.58/k, 3.73/k
r1, r2, r3, r4 = 0.98/k, 4.48/k, 0.1/k, 0.14/k
Km1, Km2, Km3, Km4 = 2.48, 3.34, 9.24, 7.24
Dt, Et =  2.12, 1.6

cAMP = np.linspace(0, 100, 100)
PDE = np.linspace(0, 100, 100)
AC = np.linspace(0, 100, 100)

ACcAMP = cAMP * AC / Km1
PDEcAMP = cAMP * PDE / Km3

plt.figure(figsize=(8,6))
plt.plot(PDE, PDEcAMP, label=r'$[PDEcAMP]$', zorder=2, linestyle='--', color='lime', linewidth=2.5, markersize=3)
plt.plot(AC, ACcAMP, label=r'$[ACcAMP]$', zorder=1, linestyle='-.', color='mediumpurple', linewidth=2.5, markersize=3)
plt.legend()
plt.xlabel(r'$[PDE]$, $[AC]$ and $[cAMP]$')
plt.ylabel(r'$[PDEcAMP]$ and $[ACcAMP]$')
plt.title(r'Plot of $[PDEcAMP]$ and $[ACcAMP]$')
plt.grid(True, linestyle=':', color='lightgray')
plt.savefig('PDEcAMP_ACcAMP_complex.png', dpi=300)
plt.show()
```
